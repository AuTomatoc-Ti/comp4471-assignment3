# Collect Submission - Zip + Generate PDF

<font color="red">**Note:** Remember to execute ALL notebooks' cells before this step. We won't be regrading missing cell outputs.</font>

Run this notebook once you have completed all the other notebooks.

It will:

* Generate a zip file of your code (`.py` and `.ipynb`) called `a3_code_submission.zip`.
* Convert all notebooks into a single PDF file called `a3_inline_submission.pdf`.

```python
import os
import shutil
import subprocess
from pathlib import Path

def run(cmd):
    print(f"$ {cmd}")
    result = subprocess.run(
        cmd,
        shell=True,
        text=True,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
    )
    if result.stdout:
        print(result.stdout)
    if result.returncode != 0:
        raise RuntimeError(f"Command failed ({result.returncode}): {cmd}")

# Resolve the working directory for both Colab and local environments
is_colab = "COLAB_GPU" in os.environ or Path("/content").exists()
folder = Path(str(FOLDERNAME)).expanduser()

if is_colab:
    # Support either a relative Drive path or an already absolute path.
    candidate = Path("/content/drive/My Drive") / folder
    workdir = candidate if candidate.exists() else folder
else:
    workdir = folder.resolve()

assert workdir.exists(), f"[!] Folder does not exist: {workdir}"
os.chdir(workdir)
print(f"Working directory: {Path.cwd()}")

# Install dependencies needed by collectSubmission.sh
if is_colab:
    run("sudo apt-get update")
    run("sudo apt-get install -y texlive-xetex texlive-fonts-recommended texlive-plain-generic pandoc")
    run("pip -q install PyPDF2")
else:
    # On macOS, MacTeX binaries are commonly here but not on PATH in notebook kernels.
    texbin = "/Library/TeX/texbin"
    if Path(texbin).exists() and texbin not in os.environ.get("PATH", ""):
        os.environ["PATH"] = f"{texbin}:{os.environ.get('PATH', '')}"
    run("pip -q install PyPDF2")

# Validate required tools before generating PDFs
required_tools = ["jupyter-nbconvert", "pandoc", "xelatex"]
missing = [tool for tool in required_tools if shutil.which(tool) is None]
assert not missing, (
    "[!] Missing required tools: " + ", ".join(missing) +
    ". On macOS, ensure pandoc is installed and add /Library/TeX/texbin to PATH (or install MacTeX)."
 )

# Generate zip + inline PDF (trim output while keeping failure code)
run("bash -o pipefail -c 'bash collectSubmission.sh 2>&1 | tail -n 40'")

# Verify expected outputs exist
for output_name in ["a3_code_submission.zip", "a3_inline_submission.pdf"]:
    output_path = Path(output_name)
    assert output_path.exists(), f"[!] Missing output file: {output_name}"
    print(f"Verified: {output_name} ({output_path.stat().st_size} bytes)")
```
